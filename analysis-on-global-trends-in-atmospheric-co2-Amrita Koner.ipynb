```python
plt.figure(figsize=(10, 6))
plt.fill_between(df['Year'], df['Uncertainty'], color='yellow', alpha=0.4)
plt.plot(df['Year'],df['Uncertainty'],label= 'Uncertainty',color='red', alpha=0.6, marker='o', linestyle='-', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Yearwise Uncertainty')
plt.legend()
plt.grid(True)
plt.tight_layout()
```

# As per above line graph we can see that there is massive down trend at the beginning in the year 1980 and at the middle it becomes little bit stable and after year 2010 again it starts increasing.

```python
plt.figure(figsize=(10, 6))
plt.fill_between(df2['Year'], df2['Uncertainty'], color='red', alpha=0.4)
plt.plot(df2['Year'], df2['Uncertainty'],label='Uncertainty',color='blue', alpha=0.6, marker='o', linestyle='-', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Yearwise Uncertainty')
plt.grid(True)
plt.tight_layout()
plt.legend()
```

```python
plt.figure(figsize=(10, 6))
plt.fill_between(df2['Year'], df2['Annual Increase'], color='black', alpha=0.4)
plt.plot(df2['Year'],df2['Annual Increase'],label='Annual Increase',color='red', alpha=0.6, marker='o', linestyle='-', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Yearwise Annual Increase')
plt.grid(True)
plt.tight_layout()
plt.legend()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df2['Year'], df2['Uncertainty'],label='Uncertainty')
plt.plot(df2['Year'],df2['Annual Increase'],label='Annual Increase')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Annual Increase and Uncertainty Over Time')
plt.legend()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df2['Annual Increase'], bins=10, alpha=0.7, label='Annual Increase')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of Annual Increase')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df2['Uncertainty'], bins=10, alpha=0.7, label='Uncertainty')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of Uncertainty')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df2['Annual Increase'], df2['Uncertainty'])
plt.xlabel('Annual Increase')
plt.ylabel('Uncertainty')
plt.title('Annual Increase vs. Uncertainty')
plt.grid(True)
plt.show()
```
